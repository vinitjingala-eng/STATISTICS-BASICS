#STATISTICS BASICS

**1.   What is the difference between descriptive statistics and inferential
statistics? Explain with examples?**

  -  Statistics is mainly divided into two types: Descriptive Statistics and Inferential Statistics. Both are used to analyze data, but they serve different purposes

     1.Descriptive Statistics:Descriptive statistics summarize and describe the main features of a dataset. It helps us understand the data we already have.It organizes, summarizes, and presents data in a meaningful way using numbers, tables, or graphs.

     Example:

     Suppose a teacher collects marks of 50 students in a class.

     Average marks = 72

     Highest marks = 95

     Lowest marks = 40

     These values describe the dataset of those 50 students only.
     No predictions are made about other students.

     2. Inferential Statistics:Inferential statistics use a sample of data to make predictions or generalizations about a larger population.It draws conclusions about a population based on sample data.

     Example:

     A researcher studies 100 voters from a city of 1,00,000 voters.

     From the sample:

     60% prefer Candidate A

     Using inferential statistics, the researcher infers that around 60% of all voters in the city may prefer Candidate A.

**2.What is sampling in statistics? Explain the differences between random
and stratified sampling.**

  -  Sampling is the process of selecting a small group of individuals (sample) from a large group (population) to collect data and draw conclusions about the entire population.

     Random Sampling:Random sampling is a method where every member of the population has an equal chance of being selected.

     Example

     If a teacher randomly selects 10 students from a class of 50 using a lottery method, each student has an equal chance of being selected.

     Stratified Sampling:Stratified sampling divides the population into smaller groups called strata based on similar characteristics (such as age, gender, department), and then samples are taken from each group.

     Example

     In a college of 1,000 students:

     600 are male

     400 are female

     If a researcher selects:

     60 males

     40 females

     This maintains the same proportion of the population.

**3.   Define mean, median, and mode. Explain why these measures of central
tendency are important?**

  -   Mean : The mean is the average value of a set of numbers. It is calculated by adding all the values and dividing by the total number of values.

      Median : The median is the middle value of a dataset when the values are arranged in ascending or descending order.

      Mode : The mode is the value that appears most frequently in a dataset.

      Measures of central tendency are statistical values that represent the central or typical value of a dataset. The three main measures are mean, median, and mode.

      -Importance of Mean, Median, and Mode

        1. Summarize Large Data

            They help represent a large dataset with one single value.

        2. Easy Comparison

             Makes it easier to compare different datasets.

        3. Understanding Data Distribution

             Shows the central point around which data values cluster.

        4. Decision Making

             Used in business, economics, research, and education to make decisions.

        5. Handling Different Data Types

             Mean is useful for numerical data.

             Median is useful when there are extreme values (outliers).

             Mode is useful for categorical data.

**4.   Explain skewness and kurtosis. What does a positive skew imply about the
data?**

  -  Skewness : Skewness is a statistical measure that describes the asymmetry of the distribution of data around its mean. In a perfectly symmetric distribution, the left and right sides are equal. When data is not symmetric, it is said to be skewed.

     Kurtosis : Kurtosis measures the peakedness or flatness of a data distribution compared to a normal distribution. It shows how heavy or light the tails of the distribution are.

     What a Positive Skew Implies About the Data

     A positive skew means:

        -The right tail of the distribution is longer.

        -Most values are smaller or moderate, but there are few very large values.

        -The mean is greater than the median and mode.

**5. Implement a Python program to compute the mean, median, and mode of
a given list of numbers.**


  numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = statistics.mean(numbers)

median_value = statistics.median(numbers)

mode_value = statistics.multimode(numbers)

print("Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

6.  Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
  
  
     list_x = [10, 20, 30, 40, 50]


    list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

cov_matrix = np.cov(list_x, list_y)
covariance = cov_matrix[0][1]

correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

7.   Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:


     data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]:

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

8.   You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and daily sales.


● Explain how you would use covariance and correlation to explore this
relationship.


● Write Python code to compute the correlation between the two lists:


  advertising_spend = [200, 250, 300, 400, 500]


  daily_sales = [2200, 2450, 2750, 3200, 4000]


Using Covariance and Correlation to Analyze the Relationship

1. Covariance

     Covariance measures how two variables change together.

     -Positive Covariance: When advertising spend increases, sales also increase.

     -Negative Covariance: When advertising spend increases, sales decrease.

     -Zero Covariance: No relationship between the variables.

     In this case, if covariance is positive, it suggests that higher advertising spending leads to higher sales.

     2. Correlation

     Correlation measures the strength and direction of the relationship between two variables.


     Its value ranges from -1 to +1.

     +1 → Perfect positive relationship

     0 → No relationship

     -1 → Perfect negative relationship

     If the correlation between advertising spend and daily sales is close to +1, it means advertising is strongly associated with increased sales

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]

print("Correlation between Advertising Spend and Daily Sales:", correlation)
```

9.  Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.?


● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.


● Write Python code to create a histogram using Matplotlib for the survey data:


survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

1. Summary Statistics to Understand the Distribution

     Before launching a new product, it is important to analyze the customer satisfaction survey scores (1–10 scale). The following statistics help understand the data:

     Mean (Average):

     -Shows the overall average satisfaction level of customers.

     -Example: If mean ≈ 7.5, it indicates customers are generally satisfied.

     Median:

     -Represents the middle value of the ordered data.

     -Useful if there are extreme scores (outliers).

     Mode:

     -Shows the most frequently occurring score.

     -Helps identify the most common customer opinion

     Standard Deviation:

     -Measures how spread out the scores are from the mean.

     -Low standard deviation → scores are similar (consistent opinions).

     -High standard deviation → opinions vary widely.

     Range:

     -Difference between the highest and lowest score.

     -Helps understand the spread of satisfaction levels

     2. Visualization


     Histogram:

     A histogram shows how frequently each score appears.

     It helps to:

     -Understand the distribution pattern

     -Identify clusters of satisfaction levels

     -Detect skewness or unusual values

     For example:

     -If most bars are around 7–9, customers are generally satisfied.

     -If many scores are below 5, there may be dissatisfaction.

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=7, edgecolor='black')

plt.title("Customer Satisfaction Survey Distribution")
plt.xlabel("Survey Scores (1-10)")
plt.ylabel("Frequency")

plt.show()
```
